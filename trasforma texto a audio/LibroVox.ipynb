```python
#librerias
import os
import tkinter as tk
from tkinter import filedialog, messagebox
import pyttsx3
from gtts import gTTS
from docx import Document
from PyPDF2 import PdfReader
from ebooklib import epub
from bs4 import BeautifulSoup
import subprocess
```

```python


# Carpetas para documentos y salida de audio
CARPETA_LIBROS = os.path.expanduser("~/Documentos/Mis Libros")
CARPETA_SALIDA = os.path.expanduser("~/Documentos/Audio Salida")

# Crear las carpetas si no existen
os.makedirs(CARPETA_LIBROS, exist_ok=True)
os.makedirs(CARPETA_SALIDA, exist_ok=True)

def convertir_azw_a_epub(ruta_archivo):
    """Convierte un archivo AZW/AZW3/MOBI a EPUB usando Calibre."""
    ruta_salida = ruta_archivo.replace(".azw", ".epub").replace(".azw3", ".epub").replace(".mobi", ".epub")

    # Comando para convertir usando ebook-convert (parte de Calibre)
    comando = ["ebook-convert", ruta_archivo, ruta_salida]

    try:
        subprocess.run(comando, check=True)
        print(f"Conversión completada: {ruta_salida}")
        return ruta_salida
    except subprocess.CalledProcessError as e:
        print(f"Error durante la conversión: {e}")
        return None

def leer_epub(ruta_archivo):
    """Extrae el contenido de un archivo EPUB."""
    texto = ""
    libro = epub.read_epub(ruta_archivo)
    for item in libro.get_items_of_type(epub.ITEM_DOCUMENT):
        soup = BeautifulSoup(item.get_content(), 'html.parser')
        texto += soup.get_text() + "\n"
    return texto

def leer_archivo(ruta_archivo):
    """Lee el contenido de archivos TXT, DOCX, PDF, EPUB, AZW, AZW3 o MOBI."""
    if ruta_archivo.endswith(".txt"):
        with open(ruta_archivo, "r", encoding="utf-8") as file:
            texto = file.read()
    elif ruta_archivo.endswith(".docx"):
        doc = Document(ruta_archivo)
        texto = "\n".join([p.text for p in doc.paragraphs])
    elif ruta_archivo.endswith(".pdf"):
        texto = ""
        with open(ruta_archivo, "rb") as file:
            reader = PdfReader(file)
            for page in reader.pages:
                texto += page.extract_text() or ""
    elif ruta_archivo.endswith(".epub"):
        texto = leer_epub(ruta_archivo)
    elif ruta_archivo.endswith((".azw", ".azw3", ".mobi")):
        epub_path = convertir_azw_a_epub(ruta_archivo)
        if epub_path:
            texto = leer_epub(epub_path)
        else:
            raise ValueError("Error al convertir el archivo Kindle.")
    else:
        raise ValueError("Formato de archivo no soportado. Usa TXT, DOCX, PDF, EPUB, AZW, AZW3 o MOBI.")
    
    return texto

def convertir_texto_a_audio(texto, idioma="es", usar_gtts=True):
    """Convierte el texto a audio usando gTTS o pyttsx3."""
    nombre_archivo = os.path.join(CARPETA_SALIDA, "audio_salida.mp3" if usar_gtts else "audio_salida.wav")

    if usar_gtts:
        tts = gTTS(text=texto, lang=idioma, slow=False)
        tts.save(nombre_archivo)
    else:
        engine = pyttsx3.init()
        engine.setProperty("rate", 150)
        engine.setProperty("volume", 0.9)
        voces = engine.getProperty("voices")
        for voz in voces:
            if idioma in voz.languages[0].decode():
                engine.setProperty("voice", voz.id)
                break
        engine.save_to_file(texto, nombre_archivo)
        engine.runAndWait()

    messagebox.showinfo("Audio generado", f"El audio ha sido guardado en: {nombre_archivo}")

def seleccionar_archivo():
    """Abre un diálogo para seleccionar el archivo desde la carpeta."""
    ruta_archivo = filedialog.askopenfilename(initialdir=CARPETA_LIBROS, 
                                              title="Selecciona un documento",
                                              filetypes=[("Documentos", "*.txt *.docx *.pdf *.epub *.azw *.azw3 *.mobi")])
    if not ruta_archivo:
        return

    try:
        texto = leer_archivo(ruta_archivo)
        if not texto.strip():
            messagebox.showwarning("Archivo vacío", "El documento seleccionado no contiene texto.")
            return

        usar_gtts = messagebox.askyesno("Seleccionar método", "¿Deseas usar gTTS (requiere internet)?")
        idioma = idioma_var.get()

        convertir_texto_a_audio(texto, idioma=idioma, usar_gtts=usar_gtts)

    except Exception as e:
        messagebox.showerror("Error", f"Se produjo un error: {str(e)}")

# Configuración de la interfaz gráfica
ventana = tk.Tk()
ventana.title("Conversor de Texto a Audio")

# Etiqueta e instrucciones
tk.Label(ventana, text="Selecciona un documento desde la carpeta 'Mis Libros':").pack(pady=10)

# Botón para seleccionar archivo
boton_seleccionar = tk.Button(ventana, text="Seleccionar documento", command=seleccionar_archivo)
boton_seleccionar.pack(pady=5)

# Selección de idioma
tk.Label(ventana, text="Selecciona el idioma para la conversión:").pack(pady=10)
idioma_var = tk.StringVar(value="es")
opciones_idioma = [("Español", "es"), ("Inglés", "en"), ("Francés", "fr")]
for texto, valor in opciones_idioma:
    tk.Radiobutton(ventana, text=texto, variable=idioma_var, value=valor).pack(anchor="w")

# Ejecutar la aplicación
ventana.mainloop()
```
